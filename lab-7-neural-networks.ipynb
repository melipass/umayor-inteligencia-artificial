#### 1.1 Funciones de activación
A continuación, definiremos tres funciones de activación a utilizar. Para que nuestro modelo calcule funciones más interesantes, necesitamos utilizar funciones de activación no lineales. Antes de comenzar, haremos una función de Python en el siguiente bloque para poder graficar las funciones de activación.



```python
x = np.linspace(-10, 10, 100)
def graficar_funcion(y):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.xticks(np.arange(-10, 11, step=5))
    plt.yticks(np.arange(-10, 11, step=.5))
    plt.xlabel("$x$")
    plt.ylabel("$g(z)$")
    plt.plot(x, y)
    plt.show()
```

##### 1.1.1 Sigmoide
La función sigmoidal es muy útil cuando estamos realizando clasificación binaria.

$$g(z)=\frac{1}{1+e^{-z}}$$

Uno de sus detalles es que mientras mayor sea z, la pendiente de la función estará más cerca de 0.

```python
sigmoide = 1/(1 + np.exp(-x))
graficar_funcion(sigmoide)
```

##### 1.1.2 tanh
Con tanh, estamos más cerca de 1 y -1 cuando ingresamos los z 

```python
tanh = np.tanh(x)
graficar_funcion(tanh)
```

##### 1.1.3 ReLU
Rectified Linear Unit.

```python
relu = np.maximum(0,x)
graficar_funcion(relu)
```
